Lets Remove the paranthesis and the names inside it inorder to get accurate Result

```python
import re
def song(data):
    hi=[]
    for i in data:
        a=re.sub("\(.*?\)","()",i)
        b=a.replace('(','').replace(')','')
        hi.append(b)
    return hi
```

Now, Lets Remove the non alphabetic values

```python
def alpha(col):
    total=[]
    for i in col:
        a=''
        for j in i:
            if j.isalpha():
                a=''.join([a, j])
        total.append(a)
    return total
```

```python
alpha(df.artist_name)
```

```python
df['artist_name']=alpha(df.artist_name)
```

```python
alpha(df.song)
```

```python
df['song']=alpha(df.song)
```

Now, Lets Create another Columns With Combined Details Of Necessary Column Details

```python
df['tags']=df['artist_name']+' '+df['song']+' '+df['duration_ms'].astype(str)
```

```python
df['tags'][0]
```

Now, Change all The Characters to Lower Case in order to get uniformilty in tags 

```python
df['tags']=df['tags'].apply(lambda x : x.lower())
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
by_name=CountVectorizer().fit_transform(df.tags)
```

```python
by_name_cs=cosine_similarity(by_name)
```
